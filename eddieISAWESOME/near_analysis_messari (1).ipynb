```python
import pandas as pd
import numpy as np
import sys, os
from datetime import datetime, timedelta
import datetime as dt

# import random_utils as ru #type: ignore
import messari #type: ignore

messari_api = 'cae559e4-f675-4c26-bb8b-545ff38594ca'
messari_client = messari.SuperMessari(messari_api)

pd.options.display.float_format = '{:,.8f}'.format
pd.set_option('display.max_rows', 100)
pd.set_option('display.max_columns', 100)
```

```python
current_time = datetime.now(dt.timezone.utc).strftime("%Y-%m-%dT%H:%M:%SZ")
n_days = 180
n_days_ago = datetime.now(dt.timezone.utc) - timedelta(days=n_days)
#n_days_ago_formatted = n_days_ago.strftime("%Y-%m-%dT%H:%M:%SZ")
n_days_ago_formatted = "2023-10-22T00:00:00Z"
```

```python
#messari_markets = pd.json_normalize(messari_markets_list)
messari_markets = pd.read_csv("messari_markets.csv")
```

```python
messari_markets["full_slug"] = messari_markets["exchangeSlug"].str.lower() + "-" + messari_markets["baseAssetSymbol"].str.lower() + "-" + messari_markets["quoteAssetSymbol"].str.lower() + "-spot"
available_exchanges = messari_markets.query("baseAssetSymbol=='NEAR' and quoteAssetSymbol.isin(['USDT', 'USD']) and exchangeSlug.isin(['binance', 'huobi', 'okx', 'mexc', 'coinbase', 'kraken', 'bybit', 'kucoin'])")
available_exchanges
```

```python
slug_id_dict = available_exchanges.set_index('full_slug')['id'].to_dict()
slug_id_dict
```

```python
consolidated_df2 = {}

for key in slug_id_dict.keys():
    consolidated_df2[key] = pd.DataFrame(messari_client.get_timeseries_by_id(slug_id_dict[key], n_days_ago_formatted, current_time, interval="1d", denomination="usd")["data"])

consolidated_df2["binance-near-usdt-spot"].loc[consolidated_df2["binance-near-usdt-spot"]["timestamp"]<1712707200, "volume"] /= 2
```

```python
consolidated_df2["binance-near-usdt-spot"]
```

```python
ohlcv_data = {"df": {}, "data": {}}
daily_volumes = []
total_average_daily_volume = []
total_average_daily_volume_usd = []
pct_limit = 0.03

for market in consolidated_df2.keys():
    ohlcv_data["df"][market] = consolidated_df2[market].copy()

    ohlcv_data["df"][market] = ohlcv_data["df"][market].sort_values(by="timestamp")

    ohlcv_data["df"][market]["average_price"] = (ohlcv_data["df"][market]["high"] + ohlcv_data["df"][market]["low"] + ohlcv_data["df"][market]["close"]) / 3
    ohlcv_data["df"][market]["usd_value"] = ohlcv_data["df"][market]["average_price"] * ohlcv_data["df"][market]["volume"]
    ohlcv_data["df"][market]["rolling_mean_volume"] = ohlcv_data["df"][market]["volume"].rolling(window=30).mean()
    ohlcv_data["df"][market]["rolling_mean_usd_value"] = ohlcv_data["df"][market]["usd_value"].rolling(window=30).mean()
    ohlcv_data["df"][market] = ohlcv_data["df"][market].dropna().reset_index(drop=True)

    df = ohlcv_data["df"][market][["timestamp", "rolling_mean_usd_value"]].copy()
    df.set_index("timestamp", inplace=True)
    df.columns = [market]
    daily_volumes.append(df)


result_df = pd.concat(daily_volumes, axis=1)
```

```python
daily_threshold_dict = {}
for i in range(0, len(result_df)):
    time = pd.to_datetime(result_df.iloc[i].name, unit="s")
    daily_threshold_dict[time] = [result_df.iloc[i].sort_values(ascending=False)[:5].mean()*0.03]

daily_threshold_df = pd.DataFrame(daily_threshold_dict).T
daily_threshold_df.to_csv("daily_trading_limit_usd.csv")
```

```python
ohlcv_data = {"df": {}, "data": {}}
total_average_daily_volume = []
total_average_daily_volume_usd = []
num_markets = len(consolidated_df2.keys())
pct_limit = 0.03

for market in consolidated_df2.keys():
    ohlcv_data["df"][market] = consolidated_df2[market].iloc[:30,:].copy()

    ohlcv_data["df"][market]["average_price"] = (ohlcv_data["df"][market]["high"] + ohlcv_data["df"][market]["low"] + ohlcv_data["df"][market]["close"]) / 3
    ohlcv_data["df"][market]["usd_value"] = ohlcv_data["df"][market]["average_price"] * ohlcv_data["df"][market]["volume"]
    #ohlcv_data["data"][market] = {"daily_average_volume": ohlcv_data["df"][market]["volume"].mean()}
    #ohlcv_data["data"][market] = {"daily_average_volume_usd": ohlcv_data["df"][market]["usd_value"].mean()}
    daily_average_volume = ohlcv_data["df"][market]["volume"].mean()
    daily_average_volume_usd = ohlcv_data["df"][market]["usd_value"].mean()
    total_average_daily_volume.append((market, daily_average_volume))
    total_average_daily_volume_usd.append((market, daily_average_volume_usd))
    print(f"{market}, daily average volume: {daily_average_volume_usd:,.2f}")

last_30d_top_exchanges = sorted(total_average_daily_volume_usd, key=lambda x: x[1], reverse=True)[:5]
print("\nTop 5 exchanges by daily average volume in the last 30 days:")
for exchange, volume in last_30d_top_exchanges:
    print(f"{exchange.split('-')[0].capitalize()}: {volume:,.2f}")
average_daily_volume = np.mean([volume for exchange, volume in last_30d_top_exchanges])
daily_trade_limit = average_daily_volume * pct_limit

print(f"\nAverage Daily Volume: {average_daily_volume:,.2f}. Daily Trade Limit: {daily_trade_limit:,.2f}")
```
